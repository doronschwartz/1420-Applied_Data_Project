```python
import pandas as pd


coal_file_path='https://github.com/doronschwartz/1420-Applied_Data_Project/blob/develop/Prelim%20Data/Global-Coal-Plant-Tracker-January-2024.xlsx?raw=true'
oil_and_gas_file_path='https://github.com/doronschwartz/1420-Applied_Data_Project/blob/develop/Prelim%20Data/Global-Oil-and-Gas-Plant-Tracker-GOGPT-August-2023.xlsx?raw=true'
steel_file_path='https://github.com/doronschwartz/1420-Applied_Data_Project/blob/develop/Prelim%20Data/Global-Steel-Plant-Tracker-2023-03-2.xlsx?raw=true'

df_coal= pd.read_excel(coal_file_path, sheet_name="Units")
df_steel=pd.read_excel(steel_file_path, sheet_name="Steel Plants")
df_oil_and_gas=pd.read_excel(oil_and_gas_file_path, sheet_name="Oil & Gas Units")
```

```python
df_coal.columns
```

```python
filtered_df_coal = df_coal[df_coal['Country']=='United States']
selected_columns = ['Country', 'Plant name','Latitude', 'Longitude', 'Location accuracy']
filtered_df_coal=filtered_df_coal[selected_columns]
filtered_df_coal.head()
```

```python
df_steel.columns
```

```python
filtered_df_steel = df_steel[df_steel['Country']=='United States']
selected_columns = ['Plant name (English)', 'Country','Coordinates', 'Coordinate accuracy',]
filtered_df_steel=filtered_df_steel[selected_columns]
filtered_df_steel.head()
```

```python
filtered_df_steel[['Latitude', 'Longitude']]= filtered_df_steel['Coordinates'].str.split(',', expand=True)
filtered_df_steel['Latitude']=pd.to_numeric(filtered_df_steel['Latitude'])
filtered_df_steel['Longitude']=pd.to_numeric(filtered_df_steel['Longitude'])
filtered_df_steel.drop(columns=['Coordinates'], inplace=True)
filtered_df_steel.head()
```

```python
# Ensure there's no leading or trailing whitespace in column names
filtered_df_steel.columns = filtered_df_steel.columns.str.strip()

# Rename columns
filtered_df_steel = filtered_df_steel.rename(columns={'Plant name (English)': 'Plant Name', 'Coordinate accuracy': 'Location Accuracy'})

# Print current column names
print(filtered_df_steel.columns)

# Reorder the columns
filtered_df_steel = filtered_df_steel[['Country', 'Plant Name', 'Latitude', 'Longitude', 'Location Accuracy']]
```

```python
df_oil_and_gas.columns
```

```python
filtered_df_oil_and_gas = df_oil_and_gas[df_oil_and_gas['Country']=='United States']
selected_columns = ['Country', 'Plant name','Latitude', 'Longitude', 'Location accuracy']
filtered_df_oil_and_gas=filtered_df_oil_and_gas[selected_columns]
filtered_df_oil_and_gas.head()
```

```python
# Rename columns in filtered_df_coal
filtered_df_coal = filtered_df_coal.rename(columns={'Plant name': 'Plant Name', 'Location accuracy': 'Location Accuracy'})

# Rename columns in filtered_df_oil_and_gas
filtered_df_oil_and_gas = filtered_df_oil_and_gas.rename(columns={'Plant name': 'Plant Name', 'Location accuracy': 'Location Accuracy'})
```

```python
num_rows_coal = len(filtered_df_coal)
num_rows_steel = len(filtered_df_steel)
num_rows_oil_and_gas = len(filtered_df_oil_and_gas)
print("Number of rows in filtered_df_coal:", num_rows_coal)
print("Number of rows in filtered_df_steel:", num_rows_steel)
print("Number of rows in filtered_df_oil_and_gas:", num_rows_oil_and_gas)
total_rows=num_rows_coal+num_rows_steel+num_rows_oil_and_gas
print("Total number of rows:", total_rows)
```

```python
filtered_df_coal['Plant Type'] = 'Coal'
filtered_df_coal.head()
```

```python
filtered_df_steel['Plant Type'] = 'Steel'
filtered_df_oil_and_gas['Plant Type']= 'Oil & Gas'
```

```python
filtered_plant_df = pd.concat([filtered_df_coal, filtered_df_steel, filtered_df_oil_and_gas], ignore_index=True)
print(filtered_plant_df)
```

```python
from geopy.distance import geodesic

# Load filtered_plant_df
# Assuming you have already imported the filtered_plant_df DataFrame

# Load uscities.csv and filter the top 50 cities by population
uscities_file_path='https://github.com/doronschwartz/1420-Applied_Data_Project/blob/develop/Prelim%20Data/uscities.csv?raw=true'
uscities_df = pd.read_csv(uscities_file_path)
top_50_cities_df = uscities_df.nlargest(50, 'population')

# Function to calculate distance between plant and city
def calculate_distance(plant_row, city_row):
    plant_location = (plant_row['Latitude'], plant_row['Longitude'])
    city_location = (city_row['lat'], city_row['lng'])
    return geodesic(plant_location, city_location).miles

# Calculate distances between each plant and the top 50 cities
for index, city_row in top_50_cities_df.iterrows():
    city_name = city_row['city']
    filtered_plant_df[f"distance_to_{city_name}"] = filtered_plant_df.apply(lambda row: calculate_distance(row, city_row), axis=1)

# Determine if any of the distances are within 60 miles
within_60_miles = filtered_plant_df[[f"distance_to_{city_row['city']}" for index, city_row in top_50_cities_df.iterrows()]].min(axis=1) <= 60

# Add a new column to filtered_plant_df indicating whether each plant is within 60 miles of any of the top 50 cities
filtered_plant_df['Within_60_Miles'] = within_60_miles

# Now filtered_plant_df contains a new column 'Within_60_Miles' indicating whether each plant is within 60 miles of any of the top 50 cities
```

```python
filtered_plant_df.head()
filtered_plant_df=filtered_plant_df[['Country', 'Plant Name', 'Latitude', 'Longitude', 'Location Accuracy', 'Plant Type', 'Within_60_Miles']]
filtered_plant_df = filtered_plant_df[filtered_plant_df['Within_60_Miles'] == False]
```

```python
filtered_plant_df.to_csv("plant_locations_top_50_cities.csv", index=False)
```
